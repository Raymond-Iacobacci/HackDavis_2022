```python
import numpy as np
import pandas as pd
files = ['01001099999 (2).csv', '01001499999 (1).csv', '01007099999.csv', '01008099999.csv', '01010099999.csv']
training_x, training_y = [], []
for i in files:
    data = pd.read_csv(i)['WDSP']
    x, y = [], []
    for i in range(0, len(data) - 100):
        training_x.append(data[i:i + 100])
        training_y.append(data[i + 100])
training_x, training_y = np.array(training_x), np.array(training_y)
```

```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, LSTM, Conv1D, Dropout, Flatten, MaxPooling1D, Reshape
aiModel = Sequential()
denseLayer = Dense(100, activation = 'relu')
aiModel.add(denseLayer)#delete dense from here
aiModel.add(Reshape((100, 1), input_shape=(100,)))
convLayer = Conv1D(filters = 50, kernel_size = 2, activation = 'relu', input_shape = (100, 1))
aiModel.add(convLayer)
aiModel.add(MaxPooling1D(pool_size = 2))
aiModel.add(LSTM(units = 8, activation = 'relu'))
aiModel.add(Dense(1))
aiModel.compile(loss = 'mean_squared_error', optimizer = 'adam')
aiModel.fit(training_x, training_y, epochs = 100, batch_size = 1)
```

```python
test_files = ['01023099999.csv']
test_x, test_y = [], []
data = pd.read_csv(test_files[0])['WDSP']
x, y = [], []
for i in range(0, len(data) - 100):
    test_x.append(data[i:i + 100])
    test_y.append(data[i + 100])
test_x, test_y = np.array(test_x), np.array(test_y)
test_y.reshape((263, 1))
print(np.shape(test_y))
print(aiModel.evaluate(test_x, test_y))
```
